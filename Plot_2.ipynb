```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for varying cache sizes
cache_sizes = [512, 1024, 2048, 4096, 8192]  # Example cache sizes

# AAT data for each replacement policy at constant associativity (4)
aat_policy1 = [4.1572 ,3.0935 , 2.4291, 1.7970 , 1.4799 ]
aat_policy2 = [4.1630  , 3.0916, 2.4278 ,1.8096 , 1.4956 ]
aat_policy3 = [4.6446, 3.4186 , 2.7080, 2.0207, 1.5907 ]
aat_policy4 = [4.5494 , 3.4060, 2.7040, 2.0434, 1.6042]

# EDP data for each replacement policy at constant associativity (4)
edp_policy1 = [626359296.0000, 355784608.0000, 333876864.0000, 213343488.0000, 167174736.0000 ]
edp_policy2 = [628660736.0000, 355551744.0000, 333927328.0000, 215484880.0000, 169618880.0000]
edp_policy3 = [ 770464448.0000, 429186880.0000, 400201120.0000, 256854032.0000, 186371904.0000]
edp_policy4 = [740967360.0000 , 427468800.0000, 399181440.0000, 262131840.0000,188849760.0000]

# Plotting AAT
plt.figure(figsize=(12, 8))
plt.plot(cache_sizes, aat_policy1, marker='o', label='LRU')
plt.plot(cache_sizes, aat_policy2, marker='s', label='Pseudo LRU')
plt.plot(cache_sizes, aat_policy3, marker='^', label='Random')
plt.plot(cache_sizes, aat_policy4, marker='d', label='Round Robin')

# Adding titles and labels for AAT plot
plt.title('Average Access Time vs. Cache Size for Different Replacement Policies\n(Associativity = 4)')
plt.xlabel('Cache Size')
plt.ylabel('Average Access Time (ns)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting EDP
plt.figure(figsize=(12, 8))
plt.plot(cache_sizes, edp_policy1, marker='o', label='LRU')
plt.plot(cache_sizes, edp_policy2, marker='s', label='Pseudo LRU')
plt.plot(cache_sizes, edp_policy3, marker='^', label='Random')
plt.plot(cache_sizes, edp_policy4, marker='d', label='Round Robin')

# Adding titles and labels for EDP plot
plt.title('Energy-Delay Product vs. Cache Size for Different Replacement Policies\n(Associativity = 4)')
plt.xlabel('Cache Size')
plt.ylabel('Energy-Delay Product')
plt.legend()
plt.grid(True)
plt.show()
```
